```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification     #설정에 따른 분류용 가상 데이터를 생성
from sklearn.cross_decomposition import PLSRegression
```

```python
X, y = make_classification(n_samples=500, n_features=500, n_informative=100, n_classes=2, random_state=2019)
fig, ax = plt.subplots(1,1,figsize=(8,8))

points, _ = PLSRegression(n_components=2).fit_transform(X,y)
ax.scatter(points[y==0,0], points[y==0,1], label="label[0]")
ax.scatter(points[y==1,0], points[y==1,1], label="label[1]")
ax.set_title("PLSRegression")
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.legend()

plt.show()
plt.close()
```
