```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np   #텍스트형식 파일의 데이터
```

```python
#dataset 생성 함수
def load_dataset(fileName):
    rows = []
    f = open(fileName, mode='r')
    
    lines = f.readlines()    #줄 단위 전체 행 읽기
    for row in lines:
        line = row.split('\t')
        row = []
        for i in line:
            row.append(float(i))
        rows.append(row)
    return np.array(rows)
```

```python
dataset = load_dataset('testSet.txt')
plt.plot(dataset[:,0], dataset[:,1], 'go')
plt.show()
```

```python
#df 생성
dataDF = pd.DataFrame(dataset, columns = ['x','y'])
dataDF.info()
```

```python
#model 생성
model = KMeans(n_clusters=4, random_state=0, algorithm='auto')
model.fit(dataDF)

pred = model.predict(dataDF)
center = model.cluster_centers_

plt.scatter(x=dataDF['x'], y=dataDF['y'], c=pred)
plt.scatter(x=center[:,0], y=center[:,1], marker='D', c='r')
plt.show()
```
